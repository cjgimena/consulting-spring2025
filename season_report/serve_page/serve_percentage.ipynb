# Serve In/Won %

```python
# imports
import pandas as pd
import numpy as np
import os
```

```python
xls = pd.ExcelFile(f'../../data/mens/Rudy Quan/combined.xlsx')

settings = xls.parse('Settings')
stats = xls.parse('Stats')
settings = settings.dropna(subset=['Host Team'])
settings = settings[['Host Team', '__source_file__']]
settings
```

```python
stats
```

```python
# merge host team and stats 
stats = stats.merge(settings, how = 'left', on = '__source_file__')
stats
```

### assuming that host team is UCLA:

```python
# total of the first serves of all the sets, assuming that host team is UCLA player
first_total = stats.loc[stats['Stat Name'] == '1st Serves']
first_serve_total = sum(first_total[['Host Set 1', 'Host Set 2', 'Host Set 3', 'Host Set 4', 'Host Set 5']].sum()) 

# total of the first serves that are in, assuming that host team is UCLA player
first_in = stats.loc[stats['Stat Name'] == '1st Serves In']
first_in_total = sum(first_in[['Host Set 1', 'Host Set 2', 'Host Set 3', 'Host Set 4', 'Host Set 5']].sum())

# total of the first serves that were won, assuming that host team is UCLA player
first_won = stats.loc[stats['Stat Name'] == '1st Serves Won']
first_won_total = sum(first_won[['Host Set 1', 'Host Set 2', 'Host Set 3', 'Host Set 4', 'Host Set 5']].sum())
```

```python
print(f'First Serve Total: {first_serve_total}')
print(f'First Serve In Total: {first_in_total}')
print(f'First Serve Won Total: {first_won_total}')
```

```python
# total of the second serves of all the sets, assuming that host team is UCLA player
second_total = stats.loc[stats['Stat Name'] == '2nd Serves']
second_serve_total = sum(second_total[['Host Set 1', 'Host Set 2', 'Host Set 3', 'Host Set 4', 'Host Set 5']].sum()) 

# total of the second serves that are in, assuming that host team is UCLA player
second_in = stats.loc[stats['Stat Name'] == '2nd Serves In']
second_in_total = sum(second_in[['Host Set 1', 'Host Set 2', 'Host Set 3', 'Host Set 4', 'Host Set 5']].sum())

# total of the second serves that were won, assuming that host team is UCLA player
second_won = stats.loc[stats['Stat Name'] == '2nd Serves Won']
second_won_total = sum(second_won[['Host Set 1', 'Host Set 2', 'Host Set 3', 'Host Set 4', 'Host Set 5']].sum())
```

```python
print(f'Second Serve Total: {second_serve_total}')
print(f'Second Serve In Total: {second_in_total}')
print(f'Second Serve Won Total: {second_won_total}')
```

```python
def add_totals(df, host, ucla_set1, ucla_set2, ucla_set3, ucla_set4, ucla_set5, opp_set1, opp_set2, opp_set3, opp_set4, opp_set5):
# def add_totals(host, df):
    if host == 'Rudy Quan':
        # return sum([ucla_set1, ucla_set2, ucla_set3, ucla_set4, ucla_set5], skipna=True)
        return np.nansum([ucla_set1, ucla_set2, ucla_set3, ucla_set4, ucla_set5])
    else:
        # return sum([opp_set1, opp_set2, opp_set3, opp_set4, opp_set5], skipna=True)
        return np.nansum([opp_set1, opp_set2, opp_set3, opp_set4, opp_set5])
```

```python
# total of the first serves for UCLA player
first_total['UCLA 1st Serves Total'] = first_total.apply(
    lambda x: add_totals(
        x, 
        x['Host Team'], x['Host Set 1'], x['Host Set 2'], x['Host Set 3'], x['Host Set 4'], x['Host Set 5'], 
        x['Guest Set 1'], x['Guest Set 2'], x['Guest Set 3'], x['Guest Set 4'], x['Guest Set 5']
        ), 
        axis=1
    )
first_serve_total2 = first_total['UCLA 1st Serves Total'].sum()

# total of the first serves in for UCLA player
first_in['UCLA 1st Serves In'] = first_in.apply(
    lambda x: add_totals(
        x, 
        x['Host Team'], x['Host Set 1'], x['Host Set 2'], x['Host Set 3'], x['Host Set 4'], x['Host Set 5'], 
        x['Guest Set 1'], x['Guest Set 2'], x['Guest Set 3'], x['Guest Set 4'], x['Guest Set 5']
        ), 
        axis=1
    )
first_in_total2  = first_in['UCLA 1st Serves In'].sum()

# total of the first serves won for UCLA player
first_won['UCLA 1st Serves Won'] = first_won.apply(
    lambda x: add_totals(
        x, 
        x['Host Team'], x['Host Set 1'], x['Host Set 2'], x['Host Set 3'], x['Host Set 4'], x['Host Set 5'], 
        x['Guest Set 1'], x['Guest Set 2'], x['Guest Set 3'], x['Guest Set 4'], x['Guest Set 5']
        ), 
        axis=1
    )
first_won_total2  = first_won['UCLA 1st Serves Won'].sum()
```

```python
print(f'First Serve Total: {first_serve_total2} and {first_serve_total}')
print(f'First Serve In Total: {first_in_total2} and {first_in_total}')
print(f'First Serve Won Total: {first_won_total2} and {first_won_total}')
```

```python
# total of the second serves for UCLA player
second_total['UCLA 2nd Serves Total'] = second_total.apply(
    lambda x: add_totals(
        x, 
        x['Host Team'], x['Host Set 1'], x['Host Set 2'], x['Host Set 3'], x['Host Set 4'], x['Host Set 5'], 
        x['Guest Set 1'], x['Guest Set 2'], x['Guest Set 3'], x['Guest Set 4'], x['Guest Set 5']
        ), 
        axis=1
    )
second_serve_total2 = second_total['UCLA 2nd Serves Total'].sum()

# total of the second serves in for UCLA player
second_in['UCLA 2nd Serves In'] = second_in.apply(
    lambda x: add_totals(
        x, 
        x['Host Team'], x['Host Set 1'], x['Host Set 2'], x['Host Set 3'], x['Host Set 4'], x['Host Set 5'], 
        x['Guest Set 1'], x['Guest Set 2'], x['Guest Set 3'], x['Guest Set 4'], x['Guest Set 5']
        ), 
        axis=1
    )
second_in_total2  = second_in['UCLA 2nd Serves In'].sum()

# total of the second serves won for UCLA player
second_won['UCLA 2nd Serves Won'] = second_won.apply(
    lambda x: add_totals(
        x, 
        x['Host Team'], x['Host Set 1'], x['Host Set 2'], x['Host Set 3'], x['Host Set 4'], x['Host Set 5'], 
        x['Guest Set 1'], x['Guest Set 2'], x['Guest Set 3'], x['Guest Set 4'], x['Guest Set 5']
        ), 
        axis=1
    )
second_won_total2  = second_won['UCLA 2nd Serves Won'].sum()
```

```python
# comparing the 2 methods.
# right now should be the same, since all of rudy's matches are home

print(f'Second Serve Total: {second_serve_total2} and {second_serve_total}')
print(f'Second Serve In Total: {second_in_total2} and {second_in_total}')
print(f'Second Serve Won Total: {second_won_total2} and {second_won_total}')
```
