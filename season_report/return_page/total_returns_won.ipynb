```python
settings = combined_data_settings.dropna(subset=['Host Team'])

returns = combined_data_shots[combined_data_shots['Type'].str.contains('return')]
returns = returns[returns['Player'] == player_name]

returns = returns.merge(settings[['__source_file__', 'Host Team']],how = 'inner', on = '__source_file__' )
returns 
```

```python
returns_total = returns.merge(combined_data_points[['__source_file__', 'Point', 'Point Winner']], how='inner', on=['Point', '__source_file__'])
returns_total
```

```python
# function to create a new column for if point winner is UCLA player

def is_return_winner(x):
    if x['Host Team'] == player_name and x['Point Winner'] == "host":
        return 1
    elif x['Host Team'] != player_name and x['Point Winner'] == "guest":
        return 1
    else:
        return 0
```

```python
returns_total['Return Winner'] = returns_total.apply(lambda x: is_return_winner(x), axis=1)

total_returns_won = sum(returns_total['Return Winner'])
total_opportunities = len(returns_total)    # total number of rows in return_totals df

percentage = (total_returns_won / total_opportunities) * 100

total_returns_won, total_opportunities, percentage
```
