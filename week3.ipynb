Create a new Column called duration in `new_df`
- using the apply function **row-rise**

```python
# new_df 
new_df = sample_df.groupby('pointNumber') [['pointStartTime']].first() 
new_df
```

```python
new_df['pointEndTime'] = sample_df.groupby('pointNumber')[['pointEndTime']].last() 
new_df
```

```python
def duration(start, end):
    duration = end - start
    return duration

new_df['duration'] = new_df.apply(lambda x: duration(x['pointStartTime'], x['pointEndTime']), axis=1) 
new_df
```

But as you notice this function could be simplified to...

```python
new_df['duration'] = new_df['pointEndTime'] - new_df['pointStartTime']
new_df
```

#### Problem 1: From the `(playsight)kaylan_usc.xlsx` dataset, classify the speed of the ball
- More than 110: **Fast**
- Less than 110 and greater than 70: **Regular**
- Less than 70: **Slow**


```python
import pandas as pd


shots_excel = pd.read_excel('data/(playsight)kaylan_usc.xlsx', sheet_name='Shots')

shots_excel.head()

def classify_speed(speed):
    if speed > 110:
        return 'Fast'
    elif 70 < speed <= 110:
        return 'Regular'
    else:
        return 'Slow'


shots_excel['Speed_Class'] = shots_excel["speed"].apply(classify_speed)

print(shots_excel)

```

#### Problem 2: From the `Shot_Visuals_KaylanBigun_NathanTrouve.csv` dataset, create a new column called "shot_depth"
- Use the image below to classify shots as deep, short, and long
- Create a function that takes the arguments `shotLocationX` and `shotLocationY` to create the outputs **short**, **deep**, and **long**

```python
from IPython.display import Image

Image(filename='images/court_dimensions.jpg', width=250)

# IN RELATION TO THIS DATASET'S COORDINATES

# Origin is at the center of the court
# 455 is the top of the court, -455 is the bottom of the court
# -210 is the left side of the court, 210 is the right side of the court
```

```python
# read in the dataset 
import pandas as pd

shot_visuals = pd.read_csv('data/Shot_Visuals_KaylanBigun_NathanTrouve.csv')

print(shot_visuals.head())
```

```python
# create function 
def classify_shot_depth(shotLocationX, shotLocationY):
    """
    Classifies the shot into 'short', 'deep', or 'long' based on coordinates.

    Parameters:
        shotLocationX (float): The X-coordinate of the shot location.
        shotLocationY (float): The Y-coordinate of the shot location.

    Returns:
        str: The classification of the shot as 'short', 'deep', or 'long'.
    """
 
    if shotLocationY > 245: 
        return 'deep'
    elif shotLocationY < -245:  
        return 'short'
    else:  
        return 'long'
```

```python
# apply function 

shot_visuals['shot_depth'] = shot_visuals.apply(
    lambda row: classify_shot_depth(row['shotLocationX'], row['shotLocationY']), axis=1
)


print(shot_visuals.head())


shot_visuals.to_csv('data/Shot_Visuals_Updated.csv', index=False)
```
