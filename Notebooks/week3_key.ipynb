### GroupBy Functions
`df.groupby()`

Create own dataset

```python
data = {
    'Name': ['Zophie', 'Alice', 'Bob', 'Charlie', 'Max', 'Milo', 'Luna', 'Bella'],
    'Age': [7, 5, 3, 2, 5, 3, 2, 5],
    'Color': ['Gray', 'Black', 'Brown', 'White', 'Black', 'Brown', 'White', 'Black'],
    'Owner': ['Charles', 'Charles', 'Cj', 'Cj', 'Cj', 'Charles', 'Cj', 'Charles'],
    'Type': ['Cat', 'Cat', 'Dog', 'Dog', 'Dog', 'Cat', 'Dog', 'Cat']
}

pets = pd.DataFrame(data)
pets
```

Count how many pets of each color each owner has

`.size()`

```python
pets_grouped = pd.DataFrame(pets.groupby(['Owner', 'Color']).size())
pets_grouped
```

### Apply Functions
`.apply()`

```python
# Use the pets dataset for this example
pets
```

```python
# Add 1 to the Age column of the pets dataset
pets['Age'] = pets['Age'].apply(lambda x: x + 1)
pets
```

```python
# same as creating your own function
def plus_one(x):
    return x + 1

pets['Age'] = pets[['Age']].apply(plus_one)
pets
```

```python
# Also use the summarizing functions (mean, max, mean)
pets['Age'].apply('mean')
```

```python
pets['Age'].apply('max')
```

```python
# But the same as this
pets['Age'].max()
```

But if you notice this code below gives us the same as our `plus_one()` function

```python
pets['Age'] + 1
```

Use case is in the conditional statements!!

```python
# create a function that classifies 5 years and younger as 'young' and everything above 'old'

def age(x):
    if x <= 5:
        return 'young'
    else:
        return 'old'
    
pets['lifeStage'] = pets['Age'].apply(age)
pets
```

Create a new Column called duration in `new_df`
- using the apply function **row-rise**

```python
new_df
```

```python
new_df_duration = new_df.copy()
new_df_duration
```

```python
def duration(start, end):
    duration = end - start
    return duration

new_df['duration'] = new_df_duration.apply(lambda x: duration(x['pointStartTime'], x['pointEndTime']), axis=1)
new_df
```

But as you notice this function could be simplified to...

```python
new_df['duration'] = new_df['pointEndTime'] - new_df['pointStartTime']
new_df
```

Use case is in the conditional statements!!
