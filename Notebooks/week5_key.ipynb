### Joining/Merging DataFrames
`.merge()`

```python
df1 = pd.DataFrame({'ID': [1, 2, 3], 
                    'Name': ['Alice', 'Bob', 'Charlie']})

df2 = pd.DataFrame({'ID': [2, 3, 4, 5], 
                    'Age': [30, 25, 40, 15]})
display(df1)
display(df2)
```

```python
# Merge the two dataframes on the ID column
df1.merge(df2, on='ID')

# Same as specifying the how argument to 'inner'
# df1.merge(df2, on='ID', how='inner')
```

```python
df1.merge(df2, on='ID', how='outer')
```

```python
# merge by left
pd.merge(df1, df2, on='ID', how='left')
```

```python
# Merge by right
pd.merge(df1, df2, on='ID', how='right')
```

#### What if I wanted to add scores to my shots sheet?

```python
def create_point_score(server, host_score, guest_score):
    host_score = str(host_score)
    guest_score = str(guest_score)

    if server == 'host':
        return host_score + "-" + guest_score
    else:
        return guest_score + "-" + host_score
    
sample_points['Point Score'] = sample_points.apply(lambda x: 
                                                   create_point_score(x['Match Server'], x['Host Game Score'], x['Guest Game Score']),
                                                   axis=1)
```

```python
sample_points
```

```python
sample_points_subset = sample_points[['Point', 'Game', 'Set', 'Point Winner', 'Detail', 'Break Point', 'Set Point', 'Point Score']].copy()
sample_points_subset
```

```python
sample_shots.merge(sample_points_subset, on = ['Point']) # notice how we have a "Game_y" and "Set_y" column
```

```python
# How to fix? Specify all the keys in your "on" argument
sample_shots_merged = sample_shots.merge(sample_points_subset, on = ['Point', 'Game', 'Set'])
sample_shots_merged # now my dataset has the additonal columns after the "Video Time" column
```
